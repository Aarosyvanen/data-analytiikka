```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('http://taanila.fi/winequality-red.csv',sep=';')
df
```

```python
X=df['density'].to_frame() #feature-matriisi X

y=df['alcohol']
```

```python
from sklearn.linear_model import LinearRegression
malli = LinearRegression()
malli.fit(X,y)
```

```python
malli.coef_
```

```python
malli.intercept_
```

```python
malli.score(X,y)
```

```python
sns.jointplot(data=df, x='density', y='alcohol', kind='reg')
```

```python
malli.predict(pd.DataFrame([0.996, 0.998, 1.000]))
```
