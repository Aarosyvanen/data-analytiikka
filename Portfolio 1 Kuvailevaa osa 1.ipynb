```python
import pandas as pd
df = pd.read_excel('http://taanila.fi/kunnat.xlsx', sheet_name='Kuntien avainluvut')
df.head()
```

```python
#uusi muuttuja
bins=[2500, 5000 , 7500 ,10000, 12500, 15000, 17500, 20000]
df['asukaslukuluokka'] = pd.cut(df['Väkiluku, 2017'],bins=bins)
df.head()
```

```python
#nimien vaihtoa
df.rename(columns = {'Alkutuotannon työpaikkojen osuus, %, 2016': 'alkutuotanto'
                     ,'Jalostuksen työpaikkojen osuus, %, 2016': 'jalostus'
                    , 'Palvelujen työpaikkojen osuus, %, 2016': 'palvelut'}, inplace = True)
df.head()
```

```python
#keskiarvot työpaikkojen osuudesta
df['keskiarvo'] = df[['alkutuotanto',
                      'jalostus',
                      'palvelut']].mean(axis=1)
df[['keskiarvo','Kunta']]
```

```python
#suodatuksia, ne joissa väkiluku yli 20000, järjestys myös vaihdettu
df[df['Väkiluku, 2017']>20000].sort_values(by='Väkiluku, 2017', ascending=False).head(20)
```
