```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('http://taanila.fi/employee.csv')
df.head()
```

```python
df.pivot_table(values='Age',index='YearsAtCompany')
```

```python
df[df['YearsAtCompany']==20].pivot_table(values = 'Age', index = 'Department').sort_values(
    'Age', ascending = False)
```

```python
df.pivot_table(values = 'Age', index = ['Department', 'YearsAtCompany'])
```

```python
df.pivot_table(values = 'Age', index = 'Department', columns = 'YearsAtCompany')
```

```python
df1 = df[df['Department']=='Sales'].pivot_table(values = 'Age', index = 'RelationshipSatisfaction', 
                                     aggfunc = [np.mean, np.median, min, max, np.std])
df1
```

```python
score4 = pd.qcut(df[df['Department']=='Sales']['Age'], 4)
df[df['Department']=='Sales'].pivot_table(values = 'Age', index = ['RelationshipSatisfaction', score4], aggfunc = len)
```
